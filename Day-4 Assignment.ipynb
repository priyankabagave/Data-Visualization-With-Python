```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 3) Plot a Pie Chart of Indian Car companies market share % values (max 6 Companies market share).
share = [20.19, 14.68, 6.25, 8.15, 14.66,21.60]
companies = ["Hyundai", "Kia", "Toyota", "Tata","Mahindra", "Maruti Suzuki"]
data = {"company": companies,"shares": share}
df = pd.DataFrame(data)
plt.pie(share, labels = companies)
plt.show()
```

```python
# 4) Plot a Bar Chart with India's covid-19 Active cases ( Max 10 States )

activecases = [119823, 10256, 65453, 37664, 38153, 2623, 20134, 1334, 5745, 1423]
states = ["Maharashtra", "Kerala", "Karnataka", "Tamil Nadu", "Andhra Pradesh", "Uttar Pradesh", "West Bengal", "Delhi", "Chhattisgarh", "Rajasthan"]
data = {"activecases": activecases,
       "states": states}
df = pd.DataFrame(data)
plt.figure(figsize=(20,10))
plt.title("Covid Active cases")
plt.xlabel("States")
plt.ylabel("Active cases")
plt.bar(states, activecases)
plt.show()
```

```python
# 5) Plot a Line Chart with India's active covid-19 case recovery (maximum 12 Months)
Months = ['May 2020','June 2020','July 2020','Aug 2020','Sept 2020','Oct 2020','Nov 2020','Dec 2020','Jan 2021','Feb 2020','Mar 2021','Apr 2021']
recovered = [158056,567536,1532135,6143019,8182882,9432075,10267263,10677710,10758619,11112056,11733514,17988537]
data = {"Months": Months,
        "recovered": recovered}
df = pd.DataFrame(data)
plt.figure(figsize=(20, 10))
plt.title("Covid-19 Case Recovery")
plt.xlabel("Months")
plt.ylabel("Recovered")
plt.plot(Months, recovered)
plt.show()
```
