```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fmri = sns.load_dataset('fmri')
fmri.head()
```

```python
#relplot
sns.relplot(x = 'timepoint', y = 'signal',data = fmri, hue = 'event',style = 'region')
```

```python
#lineplot
sns.lineplot(x = 'timepoint', y = 'signal', data = fmri, hue = 'event')
```

```python
#boxplot
sns.boxplot(x = 'region', y = 'signal', data = fmri, hue = 'event')
```
