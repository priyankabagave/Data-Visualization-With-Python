```python
#1st assignment day-1
x = np.arange(1,11)
y = x * x
plt.plot(x, y,'b*', linestyle='dashed', linewidth=2, markersize=12)
plt.xlabel("X-axis", color='blue')
plt.ylabel("Y-axis", color='blue')
plt.title("Square plot", color = 'blue')
```

```python
x = np.arange(0, 4 * np.pi,0.1)
y = np.sin(x)
plt.title("Sine Waveform", color = 'blue')
plt.xlabel("X-axis", color = 'blue')
plt.ylabel("Y-axis", color = 'blue')
plt.plot(x, y, color = 'b')
plt.show()
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*--')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
plt.subplot(2,2,4)
plt.plot(x,y,'go')
```
