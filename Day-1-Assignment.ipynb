```python
import numpy as np
import pandas as pd
import seaborn as sns

import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
from numpy.random import randn, randint, uniform, sample
```

```python
#assignment-1
x = np.arange(40,50)
y=np.arange(50,60)
plt.plot(x,y,'go--',markersize=5,linewidth=2)
plt.title("Line Graph",color='g')
plt.xlabel('X-axis',color='g')
plt.ylabel('Y-axis',color='g')
```

```python
#assignment-2
days = [1,2,3,4,5,6,7] #days of d week
sales_1 = [160,150,140,145,175,165,180] #sales of company1
sales_2 = [70,90,160,150,140,145,175]  #sales of company2
plt.plot(days,sales_1,label='Sales of Company-1')
plt.plot(days,sales_2,label='Sales of Company-2')
plt.title('Company Sales')
plt.xlabel('Days')
plt.ylabel('Sales')
plt.legend()
plt.show()
```
